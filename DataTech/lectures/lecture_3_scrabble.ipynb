```python
def read_dictionary(filename="/Users/rurikoimai/STA141b/DataTech/notebooks/data/sowpods.txt"):
    """create a list of words in the scrabble dictionary"""
    with open(filename,'r') as scrabblefile:
        scrabble_dict = [word.strip() for word in scrabblefile]
    return(scrabble_dict)
```

```python
def is_in_order(word):
    """return a true if the word is in order"""
    if word == ''.join(sorted(word)):
        return True
    else:
        print("Word is not in order")
```

```python
def find_longest_word(scrabble_dict): 
    """Returns all of the longest words in the scrabble dictionary"""
    #an empty list to store all the longest words
    words = []
    #find the length of longest word first
    longest_word =  max(scrabble_dict, key=len)
    for word in scrabble_dict:
        if len(longest_word) == len(word):
            words.append(word)
    return(words)
```

```python
find_longest_word(scrabble_dict)
```
